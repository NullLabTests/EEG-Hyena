```python
import mne
import torch
from eeg_hyena import EEGHyenaModel, preprocess_eeg

# MNE sample data (downloads automatically if not present)
sample_data_path = mne.datasets.sample.data_path()
raw_path = sample_data_path / 'MEG' / 'sample' / 'sample_audvis_raw.fif'
raw = mne.io.read_raw_fif(raw_path, preload=True)
raw.pick_types(eeg=True)  # Select EEG channels

features = preprocess_eeg(raw)

model = EEGHyenaModel(vocab_size=256, d_model=512, n_layers=1)
outputs = model(torch.tensor(features, dtype=torch.float32))
print(outputs.shape)
```
